## Extracting text from wikipedia

```python
import spacy 
from bs4 import BeautifulSoup  
import requests as rq 


link = rq.get('https://en.wikipedia.org/wiki/GPT-4')
contentRequested = 'Cha/tGPT 4 Release Date'
soup = BeautifulSoup(link.content, "html.parser")

p = soup.find_all('p')

pList = []

for paragraphs in p:
    item = paragraphs.get_text()
    pList.append(item)

text = ''.join(pList)


nlp = spacy.load("en_core_web_lg")

text2 = nlp(text)
#Saving a version of it on spacy for the nlp later

#spacy.displacy.render(text2, style = 'ent', jupyter = True)

text
```

## Setting up the Roberta Model

```python

import transformers
from transformers import AutoModelForQuestionAnswering, AutoTokenizer, pipeline

model_name = "deepset/roberta-base-squad2"

hugg = pipeline('question-answering', model=model_name, tokenizer=model_name)


#setting up the roberta model that searches for the likeliest word response given a
```

```python
QuestionAnswer = {
    'question': 'Is GPT4 released yet',
    'context': text
}

res = hugg(QuestionAnswer, top_k = 3,  max_answer_len = 1000)


#the roberta model now outputs the 3 most likeliest answers 
#https://huggingface.co/transformers/v4.6.0/_modules/transformers/pipelines/question_answering.html 

res 
```

```python


sentenceList = [sentence.text for sentence in text2.sents]


for response in res: 
    if response['score'] < 0.25: 
        #I set the confidence score to be quite low to account for multiple responsible + the it doesn't have to be super accurate since this is going to be tossed into the ChatGPT api for processing, so the goal is to have the needed info.
        pass 

    answer =  response['answer']
    recordedIndex = []

    for index, sentence in enumerate(sentenceList):
        if sentence.find(answer) != -1:
            recordedIndex.append(index)

    length = len(sentenceList)
    validSentence = []

    for index in recordedIndex: 
        # I decided to include the before and after sentence of where the answer is found to give a bit of context
        i1 = index - 1 if index > 0 else None
        i2 = index + 1 if index < length - 1 else None

        for i in [i1, index, i2]:
            if i != None:
                if len(validSentence) != 0:
                    if validSentence[-1] == sentenceList[i]:
                        pass
                print(sentenceList[i])
                validSentence.append(sentenceList[i])

result = ' '.join(validSentence)

result

#It is a little clunky because some sentence are clustered up and though of as one from the nlp. Fine tuning is the only way to deal with this, but im toooo lazy for that
#Regardless, it did highlight it is released on March 14, 2023, so it did its job!!

#Search results like these would be stored in the conversation memory of the chatbot, so while most of the information is useless, these information could be called later on to give better context so it does not have to perform the search again!
```
